##### 2. What does the `-l` flag to `ls` do? Run `ls -l /` and examine the output. What do the first 10 characters of each line mean? (Hint: `man ls`)

The `-l` flag stands for "long format" and provides detailed information about each file and directory, including file type, file mode bits (permissions), number of hard links, owner name, group name, file size, and timestamp of last modification. Ref: [long ls format](https://www.gnu.org/software/coreutils/manual/html_node/What-information-is-listed.html#index-long-ls-format)  
The first 10 characters represent the file type and permissions:
- The first character indicates the file type (e.g., `-` for regular files, `d` for directories, `l` for symbolic links).
- The next nine characters are divided into three sets of three characters each, representing the read (`r`), write (`w`), and execute (`x`) permissions for the owner, group, and others, respectively. A dash (`-`) indicates the absence of a permission. Ref: [10.1.2 What information is listed](https://www.gnu.org/software/coreutils/manual/html_node/Mode-Structure.html)

##### 4. What's the difference between `'single quotes'`, `"double quotes"`, and `$'ANSI quotes'`? Write a command that echoes a string containing a literal `$`, a `!`, and a newline character. See [Quoting](https://www.gnu.org/software/bash/manual/html_node/Quoting.html).

`''` (single quotes) preserve the literal value of each character enclosed within the quotes. No special characters are interpreted, and variables are not expanded.

`""` (double quotes) allow for variable expansion and command substitution, but still preserve most literal characters. Special characters like `$`, `` ` ``, and `\` are interpreted within double quotes.

`$''` (ANSI quotes) allow for the interpretation of backslash-escaped characters, such as `\n` for a newline or `\t` for a tab. This is useful for including special characters in a string.

````bash
echo '$SHELL'
echo "$SHELL"
echo $'Hello,\nWorld!'
````

##### 5. The shell has three standard streams: stdin (0), stdout (1), and stderr(2). Run `ls /nonexistent /tmp` and redirect stdout to one file and stderr to another. How would you redirect both to the same file? See [Redirections](https://www.gnu.org/software/bash/manual/html_node/Redirections.html).

To redirect stdout to one file and stderr to another, you can use:
```
ls /nonexistent /tmp > stdout.txt 2> stderr.txt
```
To redirect both stdout and stderr to the same file, you can use:
```
ls /nonexistent /tmp > output.txt 2>&1
```
or
```
ls /nonexistent /tmp &> output.txt
```

````bash
ls /nonexistent /tmp
````

````bash
ls /nonexistent /tmp > tmp/stdout.txt 2> tmp/stderr.txt
echo $'stdout.txt:\n'
cat tmp/stdout.txt
echo $'\nstderr.txt:\n'
cat tmp/stderr.txt
````

````bash
ls /nonexistent /tmp > tmp/output.txt 2>&1
cat tmp/output.txt
````

````bash
ls /nonexistent /tmp &> tmp/output2.txt
cat tmp/output2.txt
````

##### 7. Why does `cd` have to be built into the shell itself rather than a standalone program? (Hint: think about what a child process can and cannot affect in its parent.)

`cd` (change directory) has to be built into the shell itself because it changes the current working directory of the shell process. If `cd` were a standalone program, it would run in a separate child process, and any changes it made to the working directory would not affect the parent shell process. As a result, after the `cd` command finishes executing, the parent shell would still be in the original directory, defeating the purpose of changing directories.

##### 8. Write a script that takes a filename as an argument (`$1`) and checks whether the file exists using `test -f` or `[ -f ... ]`. It should print different messages depending on whether the file exists. See [Bash Conditional Expressions](https://www.gnu.org/software/bash/manual/html_node/Bash-Conditional-Expressions.html).

````bash
myexists() {
    if [ -f "$1" ]; then
        echo "File '$1' exists."
    else
        echo "File '$1' does not exist."
    fi
}
myexists /etc/passwd
myexists /nonexistentfile
````

##### 13. Use pipes to find the 5 most common file extensions in your home directory. (Hint: combine `find`, `grep` or `sed` or `awk`, `sort`, `uniq -c`, and `head`.)

````bash
# 方法1：grep + sed 提取扩展名
# 找有扩展名的文件 → sed提取最后的扩展名 → 统计排序
find ~ -type f | grep -E '\.[^.]+$' | sed 's/.*\.//' | sort | uniq -c | sort -nr | head -5
````

````bash
# 方法2：sed 直接提取（简洁）
# 删除最后一个点之前的所有内容，留下扩展名 → 统计排序
find ~ -type f | sed 's/.*\.//' | sort | uniq -c | sort -nr | head -5
````

````bash
# 方法3：awk -F. 提取
# 按点分割，取最后一个字段作为扩展名 → 统计排序
find ~ -type f | awk -F. '{print $NF}' | sort | uniq -c | sort -nr | head -5
````

````bash
# 方法4：rev + cut组合
# `rev` 反转字符串，`cut` 按点符取第一段，再 `rev` 反转回来 → 统计排序
find . -type f | rev | cut -d. -f1 | rev | sort | uniq -c | sort -nr | head -5
````

````bash
# 方法5：find -printf 直接提取（高效）
# 利用find的-printf指令直接输出文件名，然后用sed/awk提取扩展名
find . -type f -printf '%f\n' | sed 's/.*\.//' | sort | uniq -c | sort -nr | head -5
````
