2. What does the `-l` flag to `ls` do? Run `ls -l /` and examine the output. What do the first 10 characters of each line mean? (Hint: `man ls`)

   Answer: The `-l` flag stands for "long format" and provides detailed information about each file and directory, including file type, file mode bits (permissions), number of hard links, owner name, group name, file size, and timestamp of last modification. Ref: [long ls format](https://www.gnu.org/software/coreutils/manual/html_node/What-information-is-listed.html#index-long-ls-format)  
   The first 10 characters represent the file type and permissions:
   - The first character indicates the file type (e.g., `-` for regular files, `d` for directories, `l` for symbolic links).
   - The next nine characters are divided into three sets of three characters each, representing the read (`r`), write (`w`), and execute (`x`) permissions for the owner, group, and others, respectively. A dash (`-`) indicates the absence of a permission. Ref: [10.1.2 What information is listed](https://www.gnu.org/software/coreutils/manual/html_node/Mode-Structure.html)

3. In the command `find ~/Downloads -type f -name "*.zip" -mtime +30`, the
   `*.zip` is a "glob". What is a glob? Create a test directory with some
   files and experiment with patterns like `ls *.txt`, `ls file?.txt`, and
   `ls {a,b,c}.txt`. See [Pattern
   Matching](https://www.gnu.org/software/bash/manual/html_node/Pattern-Matching.html)
   in the Bash manual.

5. The shell has three standard streams: stdin (0), stdout (1), and stderr
   (2). Run `ls /nonexistent /tmp` and redirect stdout to one file and

6. `$?` holds the exit status of the last command (0 = success). `&&` runs

   the previous failed. Write a one-liner that creates `/tmp/mydir` only if
   it doesn't already exist. See [Exit
   Status](https://www.gnu.org/software/bash/manual/html_node/Exit-Status.html).
